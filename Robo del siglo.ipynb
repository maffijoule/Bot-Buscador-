# The bot take de list of top ordinary names and search name by name, after that save in other list the Last name, the major and the e-mail 

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
driver =webdriver.Chrome(ChromeDriverManager().install())
driver.get("http://m.udec.cl/Busqueda.php")
assert "UdeC" in driver.title
datos=[]
correos=[]
for i in nombres:
    elem = driver.find_element_by_name("palabra")
    nombre=' '+i
    elem.send_keys(nombre)
    elem.send_keys(Keys.RETURN)
    i=2
    while(True):
        try:
            path='/html/body/div[1]/div[2]/ul[1]/li['+str(i)+']'
            element = driver.find_element_by_xpath(path).text
            element=element.split('\n')
            correos.append(element)
            i+=1
        except:
            break
```

# Finally, all data go to 'excel' saving a 'RoboDelSiglo.xlsx' file.

```python
from openpyxl import load_workbook
excel_document =load_workbook('RoboDelSiglo.xlsx') #Nombre archivo, carpeta "usuario"
sheet = excel_document.active
for row in correos:
    sheet.append(row)
excel_document.save('RoboDelSiglo.xlsx')
```
